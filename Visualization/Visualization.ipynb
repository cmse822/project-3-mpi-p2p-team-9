```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
#("r1.csv")  #Blocking on same node
#("r2.csv")  #Blocking on different nodes
#("r3.csv")  #Non-Blocking on same node
#("r4.csv")  #Non-Blocking on different nodes

import pandas as pd
import matplotlib.pyplot as plt

# Define a list of CSV file names
csv_files = ["r1.csv","r2.csv", "r3.csv", "r4.csv"]

# Create an empty list to store DataFrames
dfs = []

# Load data from CSV files and calculate latency and bandwidth for each DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    df['latency'] = df[' wall_time '] / df['iterations']
    df['bandwidth'] = df['bytes'] / df[' wall_time ']/ df['iterations']
    dfs.append(df)

# Assign DataFrames to variables
df1,df2, df3,df4 = dfs

# Now df1 and df2 have the calculated latency and bandwidth
```

```python
# Plot latency vs bytes
plt.figure(figsize=(10, 5))
plt.plot(df1['bytes'], df1['latency'], label='Blocking on same node', marker='o',color ='#1f77b4')
plt.plot(df2['bytes'], df2['latency'], label='Blocking on different nodes', marker='s', color ='#ff7f0e')

plt.xlabel('Bytes')
plt.ylabel('Latency (s)')
plt.xscale('log') 
plt.yscale('log') 
plt.title('Ping Pong Benchmark')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot bandwidth vs bytes
plt.figure(figsize=(10, 5))
plt.plot(df1['bytes'], df1['bandwidth'], label='Blocking on same node', marker='o',color ='#1f77b4')
plt.plot(df2['bytes'], df2['bandwidth'], label='Blocking on different nodes', marker='s', color ='#ff7f0e')
plt.xlabel('Bytes')
plt.ylabel('Bandwidth (Bps)')
plt.xscale('log') 
plt.yscale('log') 
plt.title('Ping Pong Benchmark')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot bandwidth vs bytes
plt.figure(figsize=(10, 5))
plt.plot(df3['bytes'], df3['bandwidth'], label='Non-Blocking on same node', marker='o', color = '#2ca02c')
plt.plot(df4['bytes'], df4['bandwidth'], label='Non-Blocking on different nodes', marker='s', color ='#d62728')
plt.xlabel('Bytes')
plt.ylabel('Bandwidth (Bps)')
plt.xscale('log') 
plt.yscale('log') 
plt.title('Ping Pong Benchmark')
plt.legend()
plt.grid(True)
plt.show()
```
