```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a list of CSV file names
csv_files = ["results3.csv","results4.csv"]

# Create an empty list to store DataFrames
dfs = []

# Load data from CSV files and calculate latency and bandwidth for each DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    df['latency'] = df[' wall_time '] / (df['iterations'])
    df['bandwidth'] = df['bytes'] / df[' wall_time ']/ (df['iterations'])
    dfs.append(df)

# Assign DataFrames to variables
df1, df2 = dfs

# Now df1 and df2 have the calculated latency and bandwidth
```

```python
# Retrieve the default color cycle
default_color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

print("Default color cycle:", default_color_cycle)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
for proc in df1['n_processors'].unique():
    plt.plot(df1[df1['n_processors'] == proc]['bytes'], 
             df1[df1['n_processors'] == proc]['latency'], 
             marker='o', label=f'{proc} processors')

plt.xscale('log')  # Use logarithmic scale for better visualization
plt.yscale('log')  # Use logarithmic scale for better visualization

plt.xlabel('Data Size (bytes)')
plt.ylabel('Latency (s)')
plt.title('Blocking Ring shift')
plt.legend()
plt.grid(True)
plt.show()


# Plotting
plt.figure(figsize=(10, 6))
for proc in df1['n_processors'].unique():
    plt.plot(df1[df1['n_processors'] == proc]['bytes'], 
             df1[df1['n_processors'] == proc]['bandwidth'], 
             marker='o', label=f'{proc} processors')

plt.xscale('log')  # Use logarithmic scale for better visualization
plt.yscale('log')  # Use logarithmic scale for better visualization

plt.xlabel('Data Size (bytes)')
plt.ylabel('bandwidth (Bps)')
plt.title('Blocking Ring shift')
plt.legend()
plt.grid(True)
plt.show()
```
